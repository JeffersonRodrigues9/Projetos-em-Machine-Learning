```python
# Projeto 4 - Classificaçõa de qualidade do leite
# Neste projeto, irei analisar caracteristicas de amostras de leite para identificar a qualidade e verificar se podem
# ser itlizados na produção dos projutos alimenticios.
# O modelo de machine Learning será utilizado para fazer a classificação do leite entre as quantidades baixa, média e alta.
```

```python
import pandas as pd
```

```python
dados = pd.read_csv('/content/qualidade_leite.csv')
```

```python
# Nessa primeira etapa do projeto, o objetivo foi explorar e entender a base de dados fornecida pela indústria de laticínios, que será usada para classificar a qualidade do leite em três categorias: baixa, média e alta.

# Primeiramente, carreguei o arquivo 'qualidade_leite.csv' e verifiquei a estrutura da base de dados. Percebi que a coluna 'Qualidade' contém muitos valores nulos, o que pode impactar o modelo de machine learning. As outras colunas estão completas.

# Em seguida, analisei a distribuição dos valores na coluna 'Qualidade' e identifiquei um grande número de valores ausentes. As demais colunas numéricas estão dentro dos intervalos esperados, sem dados inconsistentes.
```

```python
dados_rotulados = dados.dropna()
```

```python
dados_rotulados
```

```python
x = dados_rotulados.drop('Qualidade', axis = 1)
y = dados_rotulados['Qualidade']
```

```python
dados_rotulados['Qualidade'].value_counts(dropna=False)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
```

```python
label_encoder.inverse_transform([0, 1, 2])
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
x_normalizado = pd.DataFrame(scaler.fit_transform(x), columns=x.columns)
x_normalizado
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(x_normalizado, y, stratify = y, random_state = 10)
```

```python
from sklearn.svm import SVC
```

```python
svm = SVC(kernel = 'linear', random_state = 10)
svm.fit(x_treino, y_treino)
```

```python
from sklearn.metrics import classification_report
```

```python
y_previsto = svm.predict(x_teste)
resultados_svm = classification_report(y_teste, y_previsto)
print(resultados_svm)
```

```python
# Nessa segunda etapa do projeto, eu realizei o pré-processamento dos dados e a construção do modelo de Machine Learning.

# 1. Tratamento de Dados Nulos
# O primeiro passo que eu tomei foi lidar com a coluna "Qualidade", que continha dados nulos. Como essa coluna é crucial para o modelo de classificação (representa a variável-alvo), não podia haver valores nulos nela. Para resolver isso, eu removi as linhas com dados nulos na coluna "Qualidade", criando o conjunto de dados rotulados.

# 2. Separação das Variáveis Explicativas e da Variável-Alvo
# Em seguida, eu separei os dados em duas partes: as variáveis explicativas (todas as colunas, exceto "Qualidade") e a variável-alvo (a coluna "Qualidade"). Armazenei as variáveis explicativas em `x` e a variável-alvo em `y`.

# Como a variável-alvo estava em formato textual (como "alta", "média", "baixa"), eu a transformei em valores numéricos utilizando o **LabelEncoder**, que converteu as categorias em números.

# 3. Normalização das Variáveis Explicativas
# Como o algoritmo **SVM (Support Vector Machine)** depende do cálculo de distâncias, foi necessário normalizar as variáveis explicativas. Eu fiz isso para garantir que todas as variáveis tivessem a mesma escala, evitando que variáveis com maiores amplitudes influenciassem mais no cálculo das distâncias. Para a normalização, utilizei o **MinMaxScaler**, que ajustou os valores das variáveis para o intervalo de 0 a 1.

# 4. Divisão dos Dados em Treinamento e Teste
# Para garantir que o modelo fosse treinado de forma eficaz e fosse capaz de fazer previsões precisas, eu dividi os dados em dois conjuntos: **treinamento** e **teste**. Usei 75% dos dados para treinamento e 25% para teste. Além disso, mantive a proporção das categorias da variável-alvo (alta, média, baixa) nos dois conjuntos, para garantir que o modelo tivesse uma distribuição equilibrada de classes.

# 5. Construção do Modelo SVM
# Com os dados preparados, eu construí um modelo de **Support Vector Machine (SVM)** com o kernel linear. Esse modelo foi treinado utilizando os dados de treinamento, ajustando os parâmetros internos para aprender os padrões presentes nas variáveis explicativas e na variável-alvo.

# 6. Avaliação do Modelo
# Após treinar o modelo, eu fiz previsões utilizando os dados de teste. Para avaliar o desempenho do modelo, calculei métricas de classificação como **precisão**, **recall**, **f1-score** e **acurácia**. Essas métricas me ajudaram a entender como o modelo estava se comportando ao fazer previsões e se estava conseguindo classificar corretamente as categorias da variável-alvo.
```

```python
dados[dados['Qualidade'].isna()]
```

```python
dados_sem_rotulo = dados[dados['Qualidade'].isna()].drop('Qualidade', axis = 1)
```

```python
dados_sem_rotulo_normalizados = pd.DataFrame(scaler.transform(dados_sem_rotulo), columns = dados_sem_rotulo.columns)
```

```python
y_previsto = svm.predict(dados_sem_rotulo_normalizados)
```

```python
y_previsto
```

```python
novo_x_treino = pd.concat([x_treino, dados_sem_rotulo_normalizados], ignore_index = True)
novo_x_treino
```

```python
novo_y_treino = pd.concat([pd.Series(y_treino), pd.Series(y_previsto)], ignore_index = True)
novo_y_treino
```

```python
pseudo_labeling = SVC(kernel='linear', random_state=10)
pseudo_labeling.fit(novo_x_treino, novo_y_treino)
```

```python
novo_y_previsto = pseudo_labeling.predict(x_teste)
resultados_pseudo_labeling = classification_report(y_teste, novo_y_previsto)
```

```python
print('SVM Supervisionado')
print(resultados_svm)
print('-'*54)
print('SVM Pseudo Labeling')
print(resultados_pseudo_labeling)
```

```python
# Terceira etapa do projeto.

#Nessa fase, apliquei a abordagem semi-supervisionada, utilizando a técnica de *pseudolabeling* para melhorar a performance do meu modelo de classificação. O objetivo foi combinar os dados rotulados e não rotulados, criando um modelo mais robusto ao aproveitar todos os dados disponíveis.

# 1. **Gerando os Pseudorótulos**
# Primeiro, analisei os dados não rotulados. A partir de um modelo supervisionado inicial, que já havia sido treinado com dados rotulados, gerei os pseudorótulos para os dados sem rótulos. Esses pseudorótulos são previsões feitas pelo modelo supervisionado para os dados não rotulados, ou seja, são rótulos "estimados", não reais.

# 2. **Concatenando os Dados**
# Após gerar os pseudorótulos, o próximo passo foi agrupar os dados rotulados com os dados não rotulados (agora com os pseudorótulos). Para isso, utilizei o método `concat()` do Pandas, que permitiu juntar as variáveis explicativas dos dados rotulados e não rotulados em uma única base de dados. Além disso, os rótulos (verdadeiros e pseudorótulos) também foram agrupados, criando uma base de dados de treinamento maior e mais diversificada.

# 3. **Treinamento do Novo Modelo**
# Com a base de dados aumentada, construí um novo modelo SVM (Support Vector Machine), agora treinado com os dados rotulados e os dados não rotulados (pseudorótulos). A ideia era comparar a performance desse modelo, que utilizava dados extras, com o modelo supervisionado original, que foi treinado apenas com dados rotulados.

# 4. **Comparativo de Desempenho**
# Após treinar o novo modelo, fiz previsões com o conjunto de dados de teste. As métricas de avaliação, como precisão, recall e F1-score, foram calculadas para comparar o modelo supervisionado original com o modelo que utilizei a abordagem semi-supervisionada (*pseudolabeling*). O modelo supervisionado foi avaliado primeiro, e, em seguida, o modelo com pseudorótulos foi comparado.

# Os resultados mostraram que o modelo supervisionado teve um desempenho ligeiramente superior em termos de acurácia
```
